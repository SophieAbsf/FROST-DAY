```python
from src.download import download_file
from src.data_loader import load_weather_data
import pandas as pd

# 📦 génération des départements (01 → 95)
departments = [f"{i:02d}" for i in range(1, 96)]

all_dfs = []

for dep in departments:
    filename = f"QUOT_departement_{dep}_periode_1950-2024_RR-T-Vent.csv.gz"

    print(f"Téléchargement : {filename}")

    try:
        filepath = download_file(filename)
        df = load_weather_data(filepath)

        df["DEPARTEMENT"] = dep  # on garde trace

        all_dfs.append(df)

    except Exception as e:
        print(f"Erreur sur {filename} : {e}")

# 🔥 fusion finale
data = pd.concat(all_dfs, ignore_index=True)

print(data.shape)
print(data.head())
```
